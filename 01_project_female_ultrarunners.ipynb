## Introduction and context 

Ultramarathons (or 'ultras') are running races of more than marathon distance, which is 26.1 miles or 42.2 km; ultrarunning is the sport of running these sorts of distances, either competitively or recreationally. A surprising finding of a recent broad study into the state of global ultrarunning (Ronto 2021) was that in events of over 195 miles, female athletes seem to to outperform male athletes. In general, differences in finish times based on gender were found to reduce for running events as distance increased, with a pattern seeming to emerge when events ranging from from 5 km to extreme ultramarathons were considered. 

I am interested in delving into the data around female participation in sport in the UK, with running, and in particular ultrarunning, as a focus. I will explore data from British ultrarunning events of different lengths, to see whether it is consistent with Ronto's findings, which considered international ultrarunning competitions. ### Social context

In 1967 the first woman, Kathrin Switzer, ran the Boston marathon; the next day the Boston Globe published shocking pictures of race officials, who disapproved of women competing in the men-only event, attempting to wrestle her off the track (Switzer 2007). Nonetheless, Switzer managed to continue, finishing in around 4 hours 20 minutes. As a result of her action, the Amateur Athletic Union (AAU) banned women from competing against men and it was not until five years later that an official women's race was established for the Boston marathon. Although mixed fields are increasingly the norm in non-elite – and some elite – running events, female participation in sport still lags behind male. In 2018 only about 18% of ultrarunners were women (Dawson 2020); although that number is gradually increasing, it helps to give some idea of the gender disparity within the sport. This tendency for women and girls to sit-out starts young, with Sport England (2020) finding that boys are more likely to be active for the recommended 60 minutes every day than girls. On a somewhat positive note, though, they found that the gap between the sexes did seem to reduce with age, with around 48% of boys and 43% of girls at secondary school meeting the aforementioned target.

Despite being in the minority, some high-profile female ultrarunning successes in recent years have captured the public imagination. For example, Jasmin Paris garnered media interest when she won the notorious 2019 Spine Race, which extends 268 miles along the Pennine Way, in a record 83 hours 12 minutes and 23 seconds, winning the event outright and setting a new record which was over 10 hours faster than the former – male – record-holder (Thomas 2020). More recently, on the other side of the world, Ruth Croft won the 2021 Ultra-Trail World Tour 102K event in Rotorua, in just over 9 hours and 20 minutes, beating her nearest male competitor by 18 minutes (Snider-McGrath 2021). Another seminal moment in female ultrarunning was when photographs were circulated of runner Sophie Power, who completed the Ultra-Trail du Mont-Blanc (UTMB) in 2018; the images showed Power at an aid station during the race, calmly feeding her 3-month-old baby and expressing breastmilk, with a male competetitor flat-out beside her looking far less composed (Pearson 2018).

### Biological context

Continuing on the theme of reproduction and lactation, it is interesting to note (though thorough exploration of this is far beyond the remit of this project) that one possible explanation of women's sucess in ultramarathons relates to inherent differences in the way that men's and women's bodies typically grow and form. Girls tend to reach puberty, and the final phase of skeletal growth, earlier than boys, and women end up having smaller, lighter skeletons and a higher proportion of body fat than men. Palmer (2009, 13) notes that 'to be too big means that seasonal food shortage[s] ... may carry more risk of death than if you are smaller', suggesting that human females evolved the dual strategies of energy-efficiency and fat-storage because it improves reproductive capacity. This is interesting to consider in the context of boundary-pushing endurance sports because the ability to keep going when the body's energy stores are depleted is key to success in ultramarathons.

## Data ethics

### Ethical considerations to do with race participants' data

Signing up to a running competition, or logging an event as a publicly visible 'running race' on Strava, involves the participants' explicit acceptance of the organisers' or platform's terms and conditions, which invariably includes giving permission to the organising institution to share run stats publicly. Since all runners whose data is visible to me will have opted in to having their run stats publically shared, the ethical considerations around the use of runners' and run-organisers' data is relatively strightforward compared to, for example, medical data, but it must be considered nonetheless.

**Ultramarathons**

Ultramarathons are, by their very nature, far more niche than shorter running events. Events are organised by a various smaller companies. An investigation into the participant data policies of some of the main organisers of UK ultras follows.

**Centurion Running** have a [nice clear section on their website](https://www.centurionrunning.com/post/data) which outlines how runners' data will be  held, processed and made publicly available, and it specifies that runners' race stats and personal details (Name, Date of Birth, Gender, Race Number) will be made publicly available on the Centurion Website 'Forever post the event'.

**Ultra X**'s [terms and conditions](https://ultra-x.co/terms-conditions/) state: 'In applying to compete in an Event, you hereby consent to Ultra X using all personal information collected from or about you in connection with the Event ... for the organisation, operation, administration and promotion of the Event. You further consent to your personal information being provided to others assisting Ultra X in connection with the Event.'

Many ultras, such as those managed by the company **Maverick Race**, use Results Base as a platform for recording and managing their results data. **Results Base**'s [privacy policy](https://resultsbase.net/privacy) mentions that athletes opt in to have their data publicly shared, but that they can opt out in which case their data will be anonymised, so anyone whose personal data is on there has accepted this condition. They state: 'Information collected from the event record will be used to produce a participant list for the event and results for the event, both may appear online.'

So – all participants in these events are opting in to having their run data, and some person information such as name, age and gender, collected and made publicly available; this is clear and explicit in the terms and conditions which they must have accepted in order to sign up to the race. I therefore feel satisfied that it is ethical to make use of this data as part of the project from this point of view.

### Permission and ethics of webscraping publicly available data from race organisers' websites

The fair use policies of some race organisers' websites make mention of webscraping and data usage. The **Deutsche Ultramarathon-Vereinigung** (DUV) website, which is what I will use for my source of ultramarathon stats, specifies that its data is subject to German copyright law, which 'applies in particular to the duplication, editing, translation, storage, processing or reproduction of content in databases or other electronic media and systems'. Luckily for us, German copyright law changed in 2018 and now very clearly allows the use of text and data mining for the purposes of education and scientific research (Stary 2018). Up to 15% of a work maybe be reproduced, stored and made publicly available for non-commercial scientific research; that limit increases to 75% for private resarch purposes. As I will be looking at a fairly small proportion of the DUV's total global database, I am well within my rights under German copyright law.

In a more general sense, it is worth considering the fact that small race organisers and results websites do not have the resources that big companies do, and their websites may therefore be more susceptiable to being compromised by aggressive webscraping of content. It is also possible that, even if the website continued to function, scraping lots of data could end up costing them lots of money, which is more problematic for a small organisation (and not very ethical either way!). In my own interests, I do not want to get blocked from a potentially useful site by their DDoS protection either. Therefore I will be mindful of this in how I implement my code and will avoid repeated webscraping from the same sites.

Overall, I feel that as long as I approach it responsibly, my accessing of publicly available race data falls under the fair-use policy of the websites I have investigated. Furthermore, race participants' personal information is publicly available because they have opted in to this as part of their event sign-up. Therefore, I feel that data access and use for this project is ethical and reasonable.

```python
# Import libraries
from bs4 import BeautifulSoup
import re
import requests
import json
import lxml
import urllib
import time
import itertools

import pandas as pd
import numpy as np
import scipy

import matplotlib.pyplot as plt
import seaborn as sns
```

### Helper functions

```python
# Define a function to get the event IDs from the hrefs

def get_event_ids(soup):
    """
    Get unique event ids from the hrefs in the html of the event-list page 
    of the statistik.d-u-v.org website.
    
    Parameters
    ----------
    soup : bs4.BeautifulSoup object
        A soup object previously generated by Beautiful Soup in combination with html.parser.

    Returns
    -------
    event_ids: list
        List  of unique event ids
    """  
    # Find the right HTML tags
    event_links = soup.select('a[href^="getresultevent.php?event="]')
    # Create empty list into which to put the event IDs
    event_ids = []
    # Iterate through event_links list
    for e in event_links: 
        # Treat the href itself as a string not a soup object
        href_string = e['href']
        # Find position in the href string of the equals sign
        equals_index = href_string.rfind('=')+1 
        # Use the position of the equals sign to slice the ID number out 
        id_only = href_string[equals_index:] 
        # Add the sliced ID numbers to the empty list
        event_ids.append(id_only) 
    # Return the list    
    return event_ids 
```

```python
# Define a function to concatenate the main part of the event URL with the event id
# Then use these to navigate to the right page and scrape the data

def make_event_url(soup):
    """
    Generate result-page url for statistik.d-u-v.org website from soup.

    Parameters
    ----------
    soup : bs4.BeautifulSoup object
        A soup object previously generated by Beautiful Soup in combination with html.parser.

    Returns
    -------
    url_list: list
        List of result-page urls for statistik.d-u-v.org website.
    """    
    # Use previous function to get event id from soup
    event_ids = get_event_ids(soup) # Use previous function to get event id from soup
    # Create empty list into which to put URLs
    url_list = []
    # Iterate through event_ids list
    for e in event_ids:
        # Concatenate the main part of the URL and the event id
        url = 'https://statistik.d-u-v.org/getresultevent.php?event=' + e 
        url_list.append(url) # Add each concatenated URL to the list of URLs
    # Return the list of URLs once it's complete
    return url_list
```

```python
# Define a function to get the soup from each event-page URL

def get_soup(url_list):
    """
    Generate soup from each page in a list of urls using Beautiful Soup and html.parser.

    Parameters
    ----------
    url_list : list
        List of result-page urls for statistik.d-u-v.org website.

    Returns
    -------
    soup_list: list of bs4.BeautifulSoup objects
        List of result-page soups for statistik.d-u-v.org website.
    """      
    # Create empty list into which to put soups
    soup_list = [] 
    # Iterate through each item in url_list
    for i,url in enumerate(url_list):
        time.sleep(5) # Don't do them all at once, wait a bit between each one
    
        # Get page data
        req = urllib.request.Request(url, headers=headers)
        with urllib.request.urlopen(req) as response:
            page_data = response.read()

        # Parse the raw HTML into a 'soup' object
        soup = BeautifulSoup(page_data, 'html.parser')
        
        soup_list.append(soup)
    return soup_list
```

```python
# Define a function to get the results data out of each page and put it into a manageable form

def parse_me_some_soup(result_soup):
    """
    Generate manageable results data from each soup object for a results page.

    Parameters
    ----------
    result_soup : bs4.BeautifulSoup object
        Result-page soup for an individual race on statistik.d-u-v.org website.

    Returns
    -------
    df: pandas DataFrame
        A dataframe of results and event details for the given race.
    """
    ####################
    # Individual results
    
    # Get the headers and add to header_list
    header_list = [h.text for h in result_soup.find_all('th')]
    
    # Find results table using HTML tags  
    table_body = result_soup.find('tbody', attrs={'id':'EvtRslt'})   
    table_rows = table_body.findAll('tr')
    
    # Go through each row of table and get the results into correct format and datatype
    all_table_data = []
    for i,table_row in enumerate(table_rows):
        table_data_for_this_row = []
        for j,col_header in enumerate(header_list):
            
            # Treat most of the data as text
            this_observation = table_row.findAll('td')[j].text.strip()
            
            # Convert incorrect 'Cat' entried to np.NaN
            if col_header=='Cat':
                if this_observation=='#NA':
                    this_observation = np.NaN
            
            # Convert 'Avg.Speed km/h' to a float
            elif col_header=='Avg.Speed km/h':
                try:
                    this_observation = float(this_observation)
                except:
                    print('Participant {}: {} is not in correct format.'.format(i, col_header))
                    this_observation = np.Nan      

            # Convert the 'Performance' and 'Age graded performance' to a single float in hours
            elif col_header in ['Performance', 'Age graded performance']:
                    # Remove the last two characters of the string: ' h'
                    this_observation = this_observation[:-2]
                    try:
                        # Split up the hours, mins and secs, and calculate as hours
                        hours, minutes, seconds = [int(x) for x in this_observation.split(":")] 
                        this_observation = hours + minutes/60 + seconds/3600
                    except ValueError:
                        try:
                            # Extract days and convert to hours
                            day_split = this_observation.split('d ')
                            days_as_hours = int(day_split[0])*24
                            # Split up the hours, mins and secs, and calculate as hours
                            hours, minutes, seconds = [int(x) for x in day_split[1].split(":")]
                            this_observation = days_as_hours + hours + minutes/60 + seconds/3600 
                        except:
                            print('Participant {}: {} is not in correct format.'.format(i, col_header))
                            this_observation = np.Nan

            # Convert 'Rank', 'Rank M/F', 'Cat. Rank' to integers
            elif col_header in ['Rank', 'Rank M/F', 'Cat. Rank']:
                try:
                    this_observation = int(this_observation)
                except:
                    print('Participant {}: {} is not in correct format.'.format(i, col_header))
                    this_observation = np.Nan 
                   
                
            table_data_for_this_row.append(this_observation)
        all_table_data.append(table_data_for_this_row)
    
    # Make a pandas df
    results_df = pd.DataFrame(all_table_data, columns=header_list)
    
    ####################
    # Event details
    
    # Define event headers list
    event_header_list = ['Event date(s)', 'Event name', \
                         'Event distance (km)', 'Event stages', \
                         'Event number of finishers']
    
    # Use the word 'Date' to locate the right info as the tags are not unique
    find_date = result_soup.find(text=re.compile('Date')) 
    data_after_date = find_date.findAllNext(text=True)
    
    # Get rid of empty items/line breaks etc.
    event_data = [ele for ele in data_after_date if ele!='\n']
    keep_index = [0,2,5,8]
    event_data_split = [event_data[:][i] for i in keep_index]

    # Split up distance and stage info if both pieces of info are included
    distance_stage = event_data_split[2]
    if '/' in distance_stage:
        try:
            distance, stage = [x for x in distance_stage.split('/')] 
        except:
            print('Distance, stage info is not in correct format.') 
            
    # Else, return the distance info as a float if possible         
    else:
        distance = distance_stage
        stage = ''
            
    # Some events have distance data in km and some in miles
    # If there's NO letter 'k' in the entry, distance is in miles and needs to be converted
    if not bool(re.search(r'km|kilometre|kilometer', distance)):
        try:
            distance = float(re.findall(r'\d+', distance)[0]) * 1.609343502101154
        except:
            print('Distance info might not be in km, and is not in correct format.')   
            distance = np.NaN            
    else:
        try:
            distance = float(re.findall(r'\d+', distance)[0])
        except:
            print('Distance info might be in km, and is not in correct format.')   
            distance = np.NaN      

    # Extract data about how many stages the event is/if it is staged
    if any(['stage' in stage]):
        num_stages = re.findall(r'\d+', stage)
        try:
            stage = int(num_stages[0])
        except:
            print('Stage info is not in correct format.')   
    else:
        stage = 1
 
    # Replace the single combined distance/stage entry with separate values 
    # for distance in km and number of stages
    event_data_split.pop(2)
    event_data_split[2:2] = distance, stage
    
    # Make a pandas df
    event_df = pd.DataFrame([event_data_split]*len(all_table_data), \
                           columns = event_header_list)
    
    ####################
    # Combine individual and event details
    
    df = pd.concat([results_df,event_df], axis=1)

    # Rename cols
    df.rename({'Surname, first name\t': 'Name', 'Original name\nSurname, first name\t': 'Name', \
               'M/F': 'Sex', 'Avg.Speed km/h': 'Mean speed (km/h)', 'Nat.': 'Nationality', \
               'Rank M/F': 'Rank (Sex)', 'Cat': 'Category', 'Cat. Rank': 'Category Rank'}, \
              axis=1, inplace=True)

    return df
```

### Finding data for UK races of over 200 km

I will begin by looking at data, including finishers' genders and times, for all UK ultramarathons of over 200 km for a given period. I have chosen to start with the longest ultramarathons as this is a useful and meaningful way of filtering the data, making a more manageable dataset to work with in the first instance. I have chosen 200 km as my initial cut-off because extremely long ultramarathons are niche at the best of times, and the last year and a half has been far from typical so there have been fewer events, I will get data for UK ultramarathons over 200 km from a longer timespan than for shorter events. I will scrape data for UK ultras of over 200 km for the years 2000-2021 inclusive. I will not get data for 2022 because the current season is not yet over and do not want to get a disproportionate amount of results for winter-spring events.

The DUV has an online database of ultrarunning events. I will use their search function to get  a webpage which lists the events over 200 km for each year that I am investigating. Then I can get the unique event ids that correspond to each event on the DUV website, which will enableme to access each results page and scrape the finishers' data.

As previously mentioned, I will scrape the data using a combination of Beautiful Soup and urllib, which will allow me to extract the publicly available race-results data from each page. I have had a look on various ultramarathon results sites and I can't find any that make the information available in a more straightforward or direct way.

```python
# Get list of all UK ultras 200 to 1000 km in 2000-2021 inclusive
urls_over_200 = []
for i in range(2000,2022):
    url = 'https://statistik.d-u-v.org/geteventlist.php?year={}' \
          '&dist=8&country=GBR&Submit.x=12&Submit.y=7&label=&surface=all&sort=1&from=200&to=1000' \
          .format(i)
    
    req = urllib.request.Request(url, headers=headers)
    
    with urllib.request.urlopen(req) as response:
        page_data = response.read()
        soup = BeautifulSoup(page_data, 'html.parser')
        urls = make_event_url(soup)
        if urls!=[]:
            urls_over_200.extend(urls)
```

### Data for UK ultras of under 200 km

The (relatively) shorter ultras have more competitors, so I don't need event data for as many years. I will just look at 2019-2021 inclusive.

```python
### Find a list of UK ultramarathons of 50–200km
distance_bins = [(50,100), (100,150), (150,200)]

# Parse the raw HTML for each page into a 'soup' object and update soup list with a loop
urls_under_200 = [[],[],[]]

for i,(x,y) in enumerate(distance_bins):
    for j in range(2019,2022):
        url = 'https://statistik.d-u-v.org/geteventlist.php?year={}' \
              '&dist=8&country=GBR&Submit.x=12&Submit.y=7&label=&surface=all&sort=1&from={}&to={}' \
              .format(j, x, y)
        
        req = urllib.request.Request(url, headers=headers)
        
        with urllib.request.urlopen(req) as response:
            page_data = response.read()
            soup = BeautifulSoup(page_data, 'html.parser')
        urls = make_event_url(soup)
        urls_under_200[i].extend(urls)
    
# Let's see how many in each category
for i,x in enumerate(urls_under_200):
    print('Events of {}-{} km in 2019-2021:  {}'.format(distance_bins[i][0], distance_bins[i][1], len(x)))
print('Total events in 2019-2021: ', sum([len(x) for x in urls_under_200]))
```

```python
# Get the soup for these events into a list of soups
urls_under_200 = [item for sublist in urls_under_200 for item in sublist]
all_soup = get_soup(urls_under_200+urls_over_200)

# Use a list comprehension to parse all the individual soups of results
ultra_df = pd.concat([parse_me_some_soup(item) for item in all_soup])

# Sort by distance and rank
ultra_df.sort_values(by=['Event distance (km)', 'Rank'], inplace=True)
ultra_df = ultra_df.reset_index(drop=True).copy()

# Uncomment to save as CSV
ultra_df.to_csv('ultra_df.csv', index=False)

ultra_df
```

## Summary of key findings/insights.

Multi-stage events tend to be outliers, with faster average times compared to straight-through events, so I have restricted data to single-stage events. When ultramarathon data is binned based on event-distance, the performance of male and female athletes gets more similar as the distance increases. In 400 km-plus events, performance of the sexes is very similar, with mean and median speeds the same for the sexes when rounded to one decimal place. About 10% of the athletes in this bracket are female, so more investigation will be needed to establish whether this might be meaningful. 

Interestingly, the common 'prestige' distances of 100 km and 100 miles have unusually shaped average-speed distributions, with a preponderance of slower runners compared to more niche distances. When 100 km and 100 mile events are removed from consideration, the histogram of athletes' average speeds is much closer to a bell curve, though it is slightly right skewed and therefore not truly Gaussian.

## Discussion and critical analysis

### Ultramarathons

Ultramarathons tend to vary a lot in terms of terrain and elevation; a shorter ultra with very tricky, technical  terrain and a lot of elevation will be completed more slowly by athletes than an easier course of the same distance, so when looking at events of a certain distance we are not necessarily comparing like with like. That is why we would expect to have quite a lot of outliers in any given distance bracket. Nevertheless, it makes sense to consider distance as one of the major defining attributes of an event.

Based on this data, we see that male athletes slightly outperform female athletes in the under 400 km events, while female athletes slightly outperform male atheletes in the 400 km-plus events. 

It is worth noting that the number of women undertaking the longest ultramarathons is extremely low. In the 400km+ distance bracket, just over 10% of competitors were women. In the longest events (500-1000km) in the UK during the years 2011-2020 there were only 5 female competitors compared to 17 male competitors. It could be suggested that there seems to be a higher barrier to entry for female athletes for these extreme events, which could suggest why the only female entrants  are elite athletes, whereas male entrants may range from elite athletes to 'normal' runners; however this is just speculation and further investigation is needed.
